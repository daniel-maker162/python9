```python
# Exercício 5: Calcular desconto
print("="*40)
print("CALCULADORA DE DESCONTO")
print("="*40)

preco_original = float(input("Digite o preço original do produto: R$ "))
percentual = float(input("Digite o percentual de desconto (%): "))

# Calcule o valor do desconto e o preço final
valor_desconto = (preco_original * percentual) / 100
preco_final = preco_original - valor_desconto

print(f"\nDesconto: R$ {valor_desconto:.2f}")
print(f"Preço final: R$ {preco_final:.2f}")
```

```python
nome = "Beltrano"

# Acessando por índice (começa em 0)
print(f"Primeira letra: {nome[0]}")   # B
print(f"Última letra: {nome[-1]}")    # o

# Fatiamento [inicio:fim] (fim não incluso)
print(f"Primeiras 3 letras: {nome[0:3]}")  # Bel
print(f"Do índice 2 ao 5: {nome[2:5]}")    # ltr

# Últimos caracteres
print(f"Últimas 3 letras: {nome[-3:]}")    # ano
```

```python
a = 10
b = 3

print(f"Soma: {a + b}")           # 13
print(f"Subtração: {a - b}")      # 7
print(f"Multiplicação: {a * b}")  # 30
print(f"Divisão: {a / b}")        # 3.333...
print(f"Divisão inteira: {a // b}") # 3
print(f"Resto: {a % b}")          # 1
print(f"Potência: {a ** b}")      # 1000

# Prioridade: parênteses > potência > * / // % > + -
resultado = (2 + 3) * 4 ** 2
print(f"(2+3)*4² = {resultado}")  # 5 * 16 = 80
```
